```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

def Data_Loader(path):
    df = pd.read_csv(path)
    df['time'] = pd.to_datetime(df['time'])
    df['lag'] = df.time.diff(1).dt.total_seconds()
    df['lag'] = np.where(df['lag']> 4800, 1, 0)
    df = df.set_index('time')
    df = df.loc[:, ['bid', 'ask', 'open', 'high', 'low', 'close', 'lag']]
    df = df.dropna()
    return df.iloc[1:, :]

texts = []
leverage = 100
for code in ['EURUSD']:
    texts += ['[switch]']
    df = Data_Loader(f"./data/{code}.csv")
    df['ticks'] = (((df['close'] - np.min(df['close'])) * leverage) * np.mean(df['close']) * 100).astype(int)
    plt.figure(figsize=[15, 3]); plt.scatter(df['ticks'].index, df['ticks'], s = 0.1)
    for i in range(df.shape[0]):
        if df['lag'][i] == 1:
            texts.append('[pause]')
            plt.plot([df['ticks'].index[i], df['ticks'].index[i]], [0, np.max(df['ticks'])], color="black")
        texts.append(df['ticks'][i])
    plt.xlabel(code); plt.ylabel('%'); plt.show()
texts = np.array(texts)
```
