```python
from markets import GlobalMarket, UtilityMarket

Market = GlobalMarket()
market_df = Market.summary()
market_df
```

```python
import numpy as np
import pandas as pd
from indicators import kelly_criterion, reverse_transform

futures_df = []
for name, code in [('NASDAQ', '^IXIC'), ('TWSE', '^TWII'), ('NYSE', '^DJI'), ('ACE', 'BTC-USD')]:
    pred, score = Market.predict(code)
    CURRENTED_PRICE, EXPECTED_PRICE, TARGET_PRICE, STATUS, POSITION = reverse_transform(market_df, code, pred[0][0], score)
    futures_df.append([name, int(CURRENTED_PRICE), int(EXPECTED_PRICE), TARGET_PRICE, STATUS, round(POSITION, 3)])
    
futures_df = pd.DataFrame(futures_df)
futures_df.columns = ['market', 'current_price', 'expected_price', 'target_price', 'status', 'position']
futures_df = futures_df.set_index('market')
futures_df
```

```python
timestamps = market_df.index
text = f"【System Info】\n Start Date: {timestamps.astype(str)[0]} \n End Date: {timestamps.astype(str)[-1]}\n\n【Option on Futures】\n"
for i in range(futures_df.shape[0]):
    status, position = futures_df.loc[futures_df.index[i], 'status'], round(futures_df.loc[futures_df.index[i], 'position'] * 100, 2)
    if futures_df.index[i] == 'NASDAQ':
        text = text + 'NASDAQ' + (f"權證/認購: {position}%" if status > 0 else f"權證/認售: {position}%" if status < 0.0 else f"趨勢不明, 待觀察") + '\n'
    elif futures_df.index[i] == 'TWSE':
        text = text + '^TWII' + (f"期貨/買權: {position}%" if status > 0 else f"期貨/賣權: {position}%" if status < 0.0 else f"趨勢不明, 待觀察") + '\n'
print(text.rstrip('\n'))
```

```python
Utility = UtilityMarket()
utility_df = Utility.summary(USD = list(market_df['USDTWD=X'])[-1])
utility_df
```

```python
html = """<html><head><style>.title_font {font-size: 24px;font-weight: bold;}</style></head><body>"""

df = Utility.predict(utility_df, futures_df, ['NASDAQ'])
html += '<p class="title_font">美股市場投資組合<br>' +  df[df['X']!=0].to_html() + "</p>"
df = Utility.predict(utility_df, futures_df, ['TWSE'])
html += '<p class="title_font">台股市場投資組合<br>' +  df[df['X']!=0].to_html() + "</p>"
df = Utility.predict(utility_df, futures_df, ['ACE'])
html += '<p class="title_font">加密貨幣市場投資組合<br>' +  df[df['X']!=0].to_html() + "</p>"
```
